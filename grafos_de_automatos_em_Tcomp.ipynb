# Plot de um automato pelo gpt

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criar o gráfico de um autômato
G = nx.DiGraph()

# Adicionar os nós (estados)
G.add_nodes_from(["q0", "q1", "q2"])

# Adicionar as transições entre os estados
G.add_edges_from([("q0", "q1", {'label': 'a'}),
                  ("q1", "q2", {'label': 'b'}),
                  ("q2", "q0", {'label': 'a'})])

# Posições dos estados no gráfico
pos = {'q0': (0, 1), 'q1': (1, 2), 'q2': (2, 1)}

# Desenhar o autômato
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=16, font_weight='bold', arrowsize=20)

# Adicionar os rótulos das transições
edge_labels = {("q0", "q1"): 'a', ("q1", "q2"): 'b', ("q2", "q0"): 'a'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14)

# Marcar o estado inicial
plt.text(-0.15, 1, "Start", fontsize=12, color='green', horizontalalignment='center')

# Mostrar o gráfico
plt.title("Autômato Exemplo")
plt.show()
```

# Novo Plot

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criar o gráfico de um autômato
G = nx.DiGraph()

# Adicionar os nós (estados)
G.add_nodes_from(["q0", "q1"])

# Adicionar as transições entre os estados
G.add_edges_from([("q0", "q0", {'label': 'a'}),
                  ("q0", "q1", {'label': 'b'}),
                  ("q1", "q0", {'label': 'a'}),
                  ("q1", "q1", {'label': 'b'})])

# Posições dos estados no gráfico
pos = {'q0': (0, 1), 'q1': (2, 1)}

# Desenhar o autômato
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=16, font_weight='bold', arrowsize=20)

# Adicionar os rótulos das transições
edge_labels = {("q0", "q0"): 'a', ("q0", "q1"): 'b', ("q1", "q0"): 'a', ("q1", "q1"): 'b'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14)

# Marcar o estado inicial e final
plt.text(-0.15, 1, "Start", fontsize=12, color='green', horizontalalignment='center')
plt.text(2.15, 1, "Final", fontsize=12, color='red', horizontalalignment='center')

# Mostrar o gráfico
plt.title("Autômato que aceita cadeias terminadas em 'b'")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criar o gráfico de um autômato
G = nx.DiGraph()

# Adicionar os nós (estados)
G.add_nodes_from(["q0", "q1"])

# Adicionar as transições entre os estados
G.add_edges_from([("q0", "q0", {'label': 'a'}),
                  ("q0", "q1", {'label': 'b'}),
                  ("q1", "q0", {'label': 'a'}),
                  ("q1", "q1", {'label': 'b'})])

# Posições dos estados no gráfico (aumentamos o espaçamento)
pos = {'q0': (0, 1), 'q1': (3, 1)}

# Desenhar o autômato com nós maiores para evitar sobreposição
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=16, font_weight='bold', arrowsize=20)

# Adicionar os rótulos das transições, ajustando as posições
edge_labels = {("q0", "q0"): 'a', ("q0", "q1"): 'b', ("q1", "q0"): 'a', ("q1", "q1"): 'b'}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=14)

# Marcar o estado inicial e final
plt.text(-0.4, 1, "Start", fontsize=12, color='green', horizontalalignment='center')
plt.text(3.35, 1, "Final", fontsize=12, color='red', horizontalalignment='center')

# Mostrar o gráfico
plt.title("Autômato que aceita cadeias terminadas em 'b'")
plt.show()
```
